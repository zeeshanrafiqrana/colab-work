```python
# Step 2: Check for Missing Values

missing_values = data.isnull().sum()
missing_percentage = (data.isnull().sum() / len(data)) * 100
print("Missing Values:\n", missing_values)
print("Missing Value Percentages:\n", missing_percentage)
```

```python
# Visualize missing values
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
sns.heatmap(data.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Step 3: Data Cleaning
```

```python
data_cleaned = data.copy()
```

```python
numerical_cols = data_cleaned.select_dtypes(include=['float64', 'int64']).columns
numerical_cols
```

```python
categorical_cols = data_cleaned.select_dtypes(include=['object']).columns
categorical_cols
```

```python
# Impute numerical columns with median
```

```python
for col in numerical_cols:
    if data_cleaned[col].isnull().sum() > 0:
        data_cleaned[col].fillna(data_cleaned[col].median(), inplace=True)
```

```python
# Impute categorical columns with mode
```

```python
for col in categorical_cols:
    if data_cleaned[col].isnull().sum() > 0:
        data_cleaned[col].fillna(data_cleaned[col].mode()[0], inplace=True)
```

```python
print("Missing values after cleaning:\n", data_cleaned.isnull().sum())
```

```python
# Step 4: Explore Data Distributions
# Understand the distribution of numerical and categorical variables.
```

```python
# Distribution of numerical columns
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(data_cleaned[col], kde=True, bins=30, color='blue')
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# Count plots for categorical columns
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(y=data_cleaned[col], palette='viridis', order=data_cleaned[col].value_counts().index)
    plt.title(f"Count Plot of {col}")
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

for col in numerical_cols:
    plt.figure(figsize=(8, 6))  # Increased figure size for better readability
    sns.kdeplot(data_cleaned[col], shade=True, color='skyblue')  # KDE plot
    plt.title(f"KDE Plot of {col}")
    plt.xlabel(col)
    plt.ylabel("Density")
    plt.show()

for col in categorical_cols:
    plt.figure(figsize=(10, 6))  # Adjusted figure size
    sns.countplot(x=data_cleaned[col], palette='viridis', order=data_cleaned[col].value_counts().index)
    plt.title(f"Count Plot of {col}")
    plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability if needed
    plt.show()
```

```python
# Step 5: Correlation Analysis
```

```python
# Correlation matrix for numerical columns
plt.figure(figsize=(10, 8))
corr_matrix = data_cleaned[numerical_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Step 6: Feature Engineering
```

```python
# Create 'Duration' feature from 'YearStart' and 'YearEnd'
if 'YearStart' in data_cleaned.columns and 'YearEnd' in data_cleaned.columns:
    data_cleaned['Duration'] = data_cleaned['YearEnd'] - data_cleaned['YearStart']
    print("Created new feature: 'Duration'")
```

```python
# Encoding categorical variables
data_encoded = pd.get_dummies(data_cleaned, columns=categorical_cols, drop_first=True)
print("Dataset after encoding categorical features:")
print(data_encoded.head())
```

```python
# Normalize numerical features
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_encoded[numerical_cols] = scaler.fit_transform(data_encoded[numerical_cols])
print("Dataset after normalization:")
print(data_encoded.head())
```
